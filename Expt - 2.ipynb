```python

import numpy as np

arr1 = np.array([1, 2, 3, 4, 5, 6, 7])
arr2 = np.array([8, 9, 10, 11, 12, 13, 14])

add = arr1 + arr2
sub = arr2 - arr1
mul = arr1 * arr2

print("Addition:", add)
print("Subtraction:", sub)
print("Multiplication:", mul)
```

```python

import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6, 7])

print("Indexing:")
print("Element at index 0:", arr[0])
print("Element at index 3:", arr[3])
print("Last element:", arr[-1])

print("Slicing:")
print("Elements from index 1 to 4:", arr[1:5])
print("First three elements:", arr[:3])
print("Elements from index 2 onwards:", arr[2:])
print("Alternate elements:", arr[::2])
```

```python
#WAP to
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

minimum = np.min(arr)
maximum = np.max(arr)

average = np.mean(arr)

print("Minimum value:", minimum)
print("Maximum value:", maximum)

avg_min = np.mean(minimum)
avg_max = np.mean(maximum)

print("Average Minimum:", avg_min)
print("Average Maximum:", avg_max)
```

```python
import numpy as np

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

conc = np.concatenate((A, B))

print("Concatenated Array:", conc)

stack = np.vstack((A, B))

print("Stacking:\n", stack)
```

```python
import numpy as np

A = np.array([10, 20, 30, 40])
B = np.array([15, 20, 25, 40])

print("A > B :", A > B)
print("A < B :", A < B)
print("A == B :", A == B)
print("A != B :", A != B)
print("A >= B :", A >= B)
print("A <= B :", A <= B)
```
