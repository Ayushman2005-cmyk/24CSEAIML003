```python
try:
  raw_x = input("Enter x value separated by space:")
  raw_y = input("Enter y value separated by space:")

  x = [float(i) for i in raw_x.split()]
  y = [float(i) for i in raw_y.split()]

  n = len(x)

  if n != len(y):
    print("Error: The values of x and y must be the same")
  elif n<2:
    print("Error: You need to least two point to form a line")
  else:
    sum_x=sum(x)
    sum_y=sum(y)
    sum_xy=sum([x[i]*y[i] for i in range(n)])
    sum_x_sq=sum(x[i]**2 for i in range(n))
    numerator_m = (n*sum_xy)-(sum_x*sum_y)
    denominator_m = (n*sum_x_sq)-(sum_x**2)
    if denominator_m == 0:
      print("Error: Vertical line detected(denominator is zero)")
    else:
      m = numerator_m/denominator_m
      c = sum_y/n-m*(sum_x/n)
      print("Slope (m):",m)
      print("\n"+"-"*30)
      print(f"Slope(m):(m:df)")
      print(f"Intercept(c):(c:df)")
      print(f"Equation: y = (m:df)x(c:df)")
      print("-"*30)
except:
  print("Error: Please enter only number separated by spaces")
```
